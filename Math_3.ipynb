При помощи линейных отображений:

    Уменьшить объект в два раза
    Повернуть на 130 градусов
    Отразить объект относительно прямой y=x (поможет матрица [[0, 1, 0], [1, 0, 0], [0, 0, 1]])



```python
import numpy as np 
import matplotlib.pyplot as plt

A = [ [-100, -100, 1], [-100, 100, 1], [100, 100, 1], [0, 0, 1], [100, -100, 1], [-100, -100, 1] ]

A = np.array(A)

x = A[:,0] 
y = A[:,1] 
plt.plot(x, y) 
plt.ylim([-200, 200]) 
plt.xlim([-200, 200]) 
```

```python
# Уменьшить объект в два раза
D = np.array([[0.5, 0,0],[0, 0.5,0],[0,0,0.5]])
A=A@D
```

```python
x = A[:,0] 
y = A[:,1] 
plt.plot(x, y) 
plt.ylim([-200, 200]) 
plt.xlim([-200, 200]) 
```

```python
# Повернуть на 130 градусов
```

```python
R = np.array([[np.cos(np.radians(130)), -np.sin(np.radians(130)),0],[np.sin(np.radians(130)), np.cos(np.radians(130)),0],[0,np.sin(np.radians(130)), np.cos(np.radians(130))]])
A= A@R
```

```python
x = A[:,0] 
y = A[:,1] 
plt.plot(x, y) 
plt.ylim([-200, 200]) 
plt.xlim([-200, 200]) 
```

```python
# Отразить объект относительно прямой y=x (поможет матрица [[0, 1, 0], [1, 0, 0], [0, 0, 1]])
```

```python
M = np.array([[0, 1, 0], [1, 0, 0], [0, 0, 1]])
A = A@M
```

```python
x = A[:,0] 
y = A[:,1] 
plt.plot(x, y) 
plt.ylim([-200, 200]) 
plt.xlim([-200, 200]) 
```

Найдите спектральное разложение матрицы:

m = np.array([[1, 2],
              [2, 3]])
              
print(m)



```python
m = np.array([[1, 2], [2, 3]])
m
```

```python
# Находим собственные значения и векторы
z, v = np.linalg.eigh(m)
z, v
```

```python
# Строим диагональную матрицу из собственных значений
D = np.diag(z)
D
```

```python
# Строим матрицу из собственных векторов и находим транспонированную матрицу
S = v.T
S
```

```python
S.T
```

```python
# Проверяем, что транспонированная и матрица из собственных векторов дают при перемножении диагональную единичную матрицу
S@S.T
```

```python
# Провереяем разложение
S.T@D@S
```
