Посчитайте частные производные функций:

$f(x,y) = 2x^2y^3 + 1/x + y^2x + 7$

$f(x,y) = x^2y - sin(xy) + cos(x^2) + 6y$

```python
import sympy as sym
```

```python
x, y = sym.symbols('x y')
```

```python
# Частные производные для первой функции
sym.diff(2*x**2*y**3+1/x+y**2*x+7, x)
```

```python
sym.diff(2*x**2*y**3+1/x+y**2*x+7, y)
```

```python
# Частные производные для второй функции
sym.diff(x**2*y-sym.sin(x*y)+sym.cos(x**2)+6*y, x)
```

```python
sym.diff(x**2*y-sym.sin(x*y)+sym.cos(x**2)+6*y, y)
```

Вычислите градиент функции:

$f(x,y) = 2x^2y^3 + 1/x + y^2x + 7$

В точке (1,2)

```python
vx = sym.diff(2*x**2*y**3+1/x+y**2*x+7, x).evalf(5, subs={x:1, y:2})
vy = sym.diff(2*x**2*y**3+1/x+y**2*x+7, y).evalf(5, subs={x:1, y:2})
grad = (vx, vy)
grad
```


Задание 3

Постройте трехмерный график функции

$f(x)=x_2^2-x_1^2 x_2 + 2 x_1 x_2$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plt_func(x1, x2):
  return x2**2-x1**2*x2+2*x1*x2
```

```python
import numpy as np
import matplotlib.cm as cm
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.ticker import LinearLocator, FormatStrFormatter
```

```python
from matplotlib import projections

fig = plt.figure(figsize=(12, 12))
ax = fig.gca(projection='3d')

X = np.arange(-15,15,0.25)
Y = np.arange(-15,15,0.25)
X, Y = np.meshgrid(X,Y)
Z = plt_func(X,Y)

ax.plot_surface(X, Y, Z, rstride=1, cstride=1,
                cmap='winter', edgecolor='none')
ax.set_xlabel('x axis')
ax.set_ylabel('y axis')
ax.set_zlabel('z axis')
ax.set_title('surface')
plt.show()
```
